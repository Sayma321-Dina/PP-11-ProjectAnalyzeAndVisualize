```python
df=pd.read_csv("student_data.csv")
df.head()
df.sample(5)
```

```python
#Answer to the q. no. 01
# List all the unique values in the "abrad_study" column

unique_values= list(set(df["abroad_study"]))
print("Unique Values: ",unique_values)
```

```python
#Answer t the q. no. 02
#Filter students who are from the "Computer Science" department and have a CGPA above 3.0

filtered_students=  df[(df["department"]=="Computer Science") & (df["cgpa"]>3.0)]
print("Filtered Students:\n",filtered_students.sample(5))
```

```python
#Answer to the q. no. 03
#Calculate the mean HSC GPA for students from the "" department.

comouter_science_sudents= df[df["department"] == "Computer Science"]
mean_hsc_gpa =computer_science_students["hsc_gpa"].mean()
mean_hsc_gpa
```

```python
#Answer to the q. no. 04
#Add a new column indicating whether students are in the "Top 10%" based on department

def top_10_percent(cgpa):
    return cgpa > 3.7

df["Top 10%"] = df.groupby("department")["cgpa"].transform(top_10_percent)
print(df.sample(10))
```

```python
#Answer to the q. no.05
# Plot a histogram of studet attendance across all deparment
import matplotlib.pyplot as plt

plt.hist(df["attendance"],bins=5,color="gray",edgecolor="black")
plt.title("Attendance Across all Department")
plt.xlabel("Department")
plt.ylabel("Attendance")
plt.show()
```
